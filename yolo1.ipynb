```python
import cv2
import numpy as np
```

```python
CONFIG_FILE="D:\Sports_Analytics\yolov3.cfg"
WEIGHTS_FILE="D:\Sports_Analytics\yolov3.weights"
net = cv2.dnn.readNet(CONFIG_FILE,WEIGHTS_FILE )
# Testing if it works
```

```python
classes=[]
with open("D:\Sports_Analytics\coco.names","r") as f:
    read=f.readlines()
for i in range(len(read)):
    classes.append(read[i].strip("\n"))
```

```python
layer_names=net.getLayerNames()
output_layers = [layer_names[i-1]for i in net.getUnconnectedOutLayers()]
```

```python
# Capture frame-by-frame
img = cv2.imread("D:\Sports_Analytics\images\image14.jpg")
```

```python
img=cv2.resize(img,None,fx=0.4,fy=0.4)
height, width, channels = img.shape
```

```python
blob = cv2.dnn.blobFromImage(img, 1 / 255.0, (416, 416),(0,0,0),True, crop=False)
```

```python
#for b in blob:
    #for n, img_blob in enumerate(b):
        #cv2.imshow(str(n),img_blob)
```

```python
#Detecting objects
net.setInput(blob)
outs = net.forward(output_layers)
print(outs)
```

```python
class_ids = []
confidences = []
boxes = []
for out in outs:
    for detection in out:
        scores = detection[5:]
        print(scores)
        class_id = np.argmax(scores)
        confidence = scores[class_id]
        if confidence > 0.25:
            center_x = int(detection[0] * width)
            center_y = int(detection[1] * height)
            w = int(detection[2] * width)
            h = int(detection[3] * height)
# Rectangle coordinates
x = int(center_x - w / 2)
y = int(center_y - h / 2)
boxes.append([x, y, w, h])
confidences.append(float(confidence))
class_ids.append(class_id)

#print(len(boxes))
```

```python
# Rectangle coordinates
x = int(center_x - w / 2)
y = int(center_y - h / 2)
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
for i in range(len(boxes)):
    x, y, w, h = boxes[i]
    label = str(classes[class_ids[i]])
    print(label)
```

```python
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
